```python
import os
from random import shuffle
```

## Load image names in list

```python
directory = "JPEGImages"
```

```python
images = [x[:-4] for x in os.listdir(directory)]
ann = [x[:-4] for x in os.listdir('Annotations')]
arr = []
for i in images:
    if i in ann:
        pass
    else:
        arr.append(i)
print(arr) # images without annotation
```

```python
image_names = os.listdir(directory)
image_names = [x[:-4] for x in image_names]
shuffle(image_names)
image_names
```

## Train test split

```python
def split(data, ratio):
    amount = int(ratio*len(data))
    train = data[:(len(data)-amount)]
    test = data[(len(data)-amount):]
    return train, test
```

```python
train, test = split(image_names, 0.2)
```

## Create text files and write

```python
def creator(data, filename):
    file = open(filename, "w")
    with open(filename, 'a') as the_file:
        for i in data:
            the_file.write(i+'\n')
    file.close() 
```

```python
import os    
save_dir = "ImageSets/Main/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
creator(train, save_dir+"train.txt")
creator(test, save_dir+"test.txt")
```
